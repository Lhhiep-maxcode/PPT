```python
import math
```

```python
def newton(p0, tol, iternums=10000):
    p = p0 - f(p0) / df(p0)
    print(p)
    if abs(p - p0) < tol or iternums == 1:
        return p
    return newton(p, tol, iternums - 1)
```

```python
f = lambda x: x ** 3 - 2 * x ** 2 - 5
df = lambda x: 3 * x ** 2 - 4 * x
tol = 10 ** -4
newton(2, tol)
```
