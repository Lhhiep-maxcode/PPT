```python
from decimal import Decimal, getcontext
```

```python
getcontext().prec = 29
```

```python
def partial_sum(n):
    sum = 0
    for i in range(1, 1 + n):
        sum += Decimal(2) ** Decimal(-4 * i)
    return Decimal(3 / 2) * sum
```

```python
n = 1

for i in range(30):
    psum = partial_sum(i + 1)
    print(f"Loop {i} - Approximation: {psum}")

    
```
