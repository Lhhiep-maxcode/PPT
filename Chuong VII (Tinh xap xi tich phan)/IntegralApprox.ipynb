```python
import math
```

```python
def TrapezoidalRule(f, start, end, n):
    h = (end - start) / n # Step size
    integral = 0 # The result

    integral += (f(start) + f(end)) * h / 2.0
    for i in range(1, n):
        x = start + i * h # Calculate x coordinate
        integral += f(x) * h

    return integral

def MidpointRule(f, start, end , n):
    h = (end - start) / n # Step size
    integral = 0

    for i in range(1, n, 2):
        x = start + i * h
        integral += f(x) * 2 * h

    return integral

def SimpsonRule(f, start, end, n):
    h = (end - start) / n # Step size
    integral = 0

    for i in range(1, n, 2):
        x = start + i * h
        prev_x = x - h
        next_x = x + h
        integral += f(prev_x) + 4 * f(x) + f(next_x)

    integral *= h/3

    return integral
```

```python
f = lambda x: x**4
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 0.5, 1, 10)
print(ans)
```

```python
f = lambda x: 2 / (x - 4)
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 0, 0.5, 10)
print(ans)
```

```python
f = lambda x: x**2 * math.log(x)
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 1, 1.5, 10)
print(ans)
```

```python
f = lambda x: x**2 * math.exp(-x)
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 0, 1, 20)
print(ans)
```

```python
f = lambda x: 2 * x / (x**2 - 4)
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 1, 1.6, 20)
print(ans)
```

```python
f = lambda x: 2 / (x**2 - 4)
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 0, 0.35, 20)
print(ans)
```

```python
f = lambda x: x * math.sin(x)
# the number of chunk (parameter n) should be even 
ans = SimpsonRule(f, 0, math.pi / 4, 20)
print(ans)
```
