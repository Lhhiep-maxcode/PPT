```python
import math
import numpy as np
```

```python
def gaussSeidel(A, x, b, epsilon=1e-3, iterNums=10):
    print("\nGauss-Seidel Method:")
    A = np.array(A)
    x = np.array(x)
    b = np.array(b)
    xSize = len(x)
    for i in range(iterNums):
        x_old = np.array(x)
        for j in range(xSize):
            x[j] = 1.0 / A[j, j] * (b[j, 0] - (A[j] @ x - A[j, j] * x[j, 0]))
        print(x.reshape((1, -1)))
        if np.max(np.abs(x_old - x)) < epsilon:
            print(np.max(np.abs(x_old - x)))
            return list(x)

    return list(x)
```

```python
A = [[10.,    1.,    1.],
     [ 1.,   10.,    1.],
     [ 1.,    1.,   10.]]

x = [[6.],
     [6.],
     [6.]]

b = [[6.],
     [6.],
     [6.]]


x = gaussSeidel(A, x, b)
print()
for i in range(len(x)):
    print(f"x{i + 1} = {x[i]}")
```
