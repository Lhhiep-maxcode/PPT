```python
import math
import numpy as np
```

```python
def rowInterchange(A, m, i):
    tmp = np.copy(A[i])
    A[i] = A[m]
    A[m] = tmp

def backSubstitute(A, x):
    sysSize = len(A)
    varSize = len(x)
    for i in reversed(range(sysSize)):
        x[i] = 1 / A[i, i] * (A[i, varSize] - A[i, i + 1:varSize] @ x[i + 1:varSize])


def gaussElimination(A, x, b):
    print("\nGauss Elimination Method:")
    A = np.array(A)
    x = np.array(x)
    b = np.array(b)
    A = np.concatenate((A, b), axis=1)
    unique_sol = True
    sysSize = len(A)
    for i in range(sysSize):
        m = i
        for k in range(i + 1, sysSize):
            if abs(A[k, i]) > abs(A[m, i]):
                m = k
        if A[m, i] == 0:
            unique_sol = False
            continue
        rowInterchange(A, m, i)
        for k in range(i + 1, sysSize):
            m_ki = A[k, i] / A[i, i]
            A[k] = A[k] - m_ki * A[i]
    if not unique_sol:
        if (A[sysSize - 1] == np.array([0] * (sysSize + 1))).all():
            return "Infinite Solution"
        else:
            return "No solution"
    backSubstitute(A, x)
    return list(x)
```

```python
A = [[1., -1., 3.],
     [3., -3., 1.],
     [1., 1., 0.]]

x = [[0.],
     [0.],
     [0.]]

b = [[2.],
     [-1.],
     [3.]]
x = gaussElimination(A, x, b)
if type(x) == str:
    print(x)
else:
     for i in range(len(x)):
          print(f"x{i + 1} = {x[i]}")
```
