```python
import scipy
import numpy as np
import scipy.linalg
```

```python
def LU_factorization(A: np.ndarray, b: np.ndarray) -> np.ndarray:
    """
    Solve Ax = b precisely using LU factorization 
    => LUx = b 
    => Ux = inv(L) * b
    Return an array of x_1, x_2, ..., x_n
    """
    _, l, u = scipy.linalg.lu(A)
    print(f"Ma trận đường chéo dưới: L = ")
    print(l)

    print()
    print()
    
    print(f"Ma trận đường chéo trên: U = ")
    print(u)
    
    b = scipy.linalg.inv(l) @ b
    x = scipy.linalg.inv(u) @ b
    print()
    return x.flatten()
```

```python
A = [[ 3.,     9.,     6.],
     [18.,    48.,    39.],
     [ 9.,   -27.,    42.]]

x = [[0.],
     [0.],
     [0.]]

b = [[     4.6],
     [    27.2],
     [     9.0]]

x = LU_factorization(A, b)
print("Nghiệm của hệ PT:")
for i in range(len(x)):
     print(f"x{i + 1} = {x[i]}")
```
