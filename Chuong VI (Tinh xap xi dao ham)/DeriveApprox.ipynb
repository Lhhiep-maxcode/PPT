```python
def midpoint_rule_3point(f, target, h):
    d =  (f(round(target+h, 4)) - f(round(target-h, 4))) / (2.0*h)
    return d

def left_endpoint_rule_3point(f, target, h):
    d =  (-3*f(round(target, 4)) + 4*f(round(target+h, 4)) - f(round(target+2*h, 4))) / (2.0*h)
    return d

def right_endpoint_rule_3point(f, target, h):
    d = (3*f(round(target, 4)) - 4*f(round(target-h, 4)) + f(round(target-2*h, 4))) / (2.0*h)
    return d

def midpoint_rule_5point(f, target, h):
    d = (f(round(target-2*h, 4)) - 8*f(round(target-h, 4)) + 8*f(round(target+h, 4)) - f(round(target+2*h, 4))) / (12*h)
    return d

def endpoint_rule_5point(f, target, h):
    d = (-25*f(round(target, 4)) + 48*f(round(target+h, 4)) - 36*f(round(target+2*h, 4)) + 16*f(round(target+3*h, 4)) - 3*f(round(target+4*h, 4))) / (12.0*h)
    return d

def snd_derivative(f, target, h):
    d = h**(-2) * (f(round(target-h, 4)) - 2*f(round(target, 4)) + f(round(target+h)))
    return d
```

```python
def round_to_four_digits(value):
    if value == 0:
        return 0
    else:
        return round(value, 4 - int(np.floor(np.log10(abs(value)))) - 1)
```
