```python
import math
import numpy as np
```

```python
def midpoint_method(y0, x0, x_end, df, h):
    print("Midpoint method: ")
    while x0 < x_end - 1e-9:
        y_tmp = y0 + h / 2 * df(x0, y0)
        y0 = y0 + h * df(x0 + h/2, y_tmp)
        x0 = x0 + h
        print(f"y({round(x0, 4)}) = {y0} - actual value = {y(x0)}")
    print()
```

```python
dy = lambda x, y: x * math.exp(3 * x) - 2 * y
y = lambda x: 1/5 * x * math.exp(3 * x) - 1 / 25 * math.exp(3 * x) + 1 / 25 * math.exp(-2 * x)
midpoint_method(0, 0, 1, dy, 0.5)
```

```python
dy = lambda x, y: 1 + (x - y)**2
y = lambda x: x + 1 / (1 - x)
midpoint_method(1, 2, 3, dy, 0.5)
```

```python
dy = lambda x, y: 1 + y / x
y = lambda x: x * math.log(x) + 2 * x
midpoint_method(2, 1, 2, dy, 0.25)
```

```python
dy = lambda x, y: math.cos(2 * x) + math.sin(3 * x)
y = lambda x: 1 / 2 * math.sin(2 * x) - 1 / 3 * math.cos(3 * x) + 4 / 3
midpoint_method(1, 0, 1, dy, 0.25)
```
