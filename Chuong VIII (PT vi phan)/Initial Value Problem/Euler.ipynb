```python
import math
```

```python
def euler_method(y0, x0, x_end, df, h):
    print("Euler Method:")
    while x0 < x_end - 1e-9:
        y0 = y0 + h * df(x0, y0)
        x0 = x0 + h
        print(f"y({round(x0, 4)}) = ", y0)
    print()
```

```python
df = lambda x, y: y / x - (y / x)**2
euler_method(1, 1, 2, df, 0.1)
```

```python
df = lambda x, y: 1 + y / x + (y / x)**2
euler_method(0, 1, 3, df, 0.2)
```

```python
df = lambda x, y: - (y + 1) * (y + 3)
euler_method(-2, 0, 2, df, 0.2)
```

```python
df = lambda x, y: -5 * y + 5 * x**2 + 2 * x
euler_method(1/3, 0, 1, df, 0.1)
```
