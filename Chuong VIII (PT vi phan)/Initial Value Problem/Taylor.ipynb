```python
import math
import numpy as np
```

```python
def taylor_series_method(y0: float, x0: float, x_end: float, df, h: float):
    print("Taylor Series Method:")
    i = 0
    while x0 < x_end:
        print("# i = ", i, ":", sep="", end=" ")
        y0 = y0 + h * df(x0, y0) 
        x0 = x0 + h
        print(f"y({round(x0, 4)}) = ", y0)
        i += 1

    print()
```

```python
df = lambda x, y: y / x - (y / x)**2
taylor_series_method(1, 1, 1.2, df, 0.1)
```

```python
df = lambda x, y: math.sin(x) + math.exp(-x)
taylor_series_method(0, 0, 1, df, 0.5)
```

```python
df = lambda x, y: (y**2 + y) / x
taylor_series_method(-2, 1, 3, df, 0.5)
```

```python
df = lambda x, y: -x * y + 4 * x * y**-1
taylor_series_method(1, 0, 1, df, 0.25)
```
