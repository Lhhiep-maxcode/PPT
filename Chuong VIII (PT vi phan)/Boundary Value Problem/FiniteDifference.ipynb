```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# h = 0.1

### Config here

```python
N=10
h=1/N
x=np.linspace(0,1,N+1)
```

```python
y=np.zeros((N+1))
# Boundary Condition
y[0]=1
y[N]=math.exp(-10)
```

```python
# Ay = b

A=np.zeros((N-1,N-1))
# Diagonal
for i in range (0,N-1):
    A[i,i] = -2 / (h * h * 100)

for i in range (0,N-2):           
    A[i+1,i] = (1 / (h * h * 100) - 1)
    A[i,i+1] = (1 / (h * h * 100))
```

```python
b=np.zeros(N-1)

# Boundary Condition
b[0] = y[0] * (1 - 1 / (100 * h * h))
b[N-2]= y[N] * (1 - 1 / (100 * h * h))
```

### Calculate

```python
invA=np.linalg.inv(A)
y[1:N]=np.dot(invA,b)
```

```python
fig = plt.figure(figsize=(8,4))

plt.plot(x,y,'o',label='Finite Difference')
plt.plot(x,np.exp(-10 * x),'k:',label='Truth')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```
