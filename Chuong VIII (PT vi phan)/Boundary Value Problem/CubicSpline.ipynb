```python
import numpy as np
from scipy.interpolate import CubicSpline
import math
import matplotlib.pyplot as plt
```

```python
y = lambda x: x**2 - x
x_start = 0
x_end = 1
h = 0.1
x = np.linspace(0, 1, int((x_end - x_start) / h))
cs = CubicSpline(x=x, y=y(x))
plt.plot(x, cs(x), ':o', label='Cubic Spline', color='green')
x = np.linspace(0, 1, 100)
plt.plot(x, y(x), label='truth', color='red')
plt.legend()
```

```python
y = lambda x: (x - 1) * (np.exp(-x) - 1)
x_start = 0
x_end = 1
h = 0.1
x = np.linspace(0, 1, int((x_end - x_start) / h))
cs = CubicSpline(x=x, y=y(x))
plt.plot(x, cs(x), ':o', label='Cubic Spline', color='green')
x = np.linspace(0, 1, 100)
plt.plot(x, y(x), label='truth', color='red')
plt.legend()
```

```python
y = lambda x: x * (np.exp(x) - np.e)
x_start = 0
x_end = 1
h = 0.05
x = np.linspace(0, 1, int((x_end - x_start) / h))
cs = CubicSpline(x=x, y=y(x))
plt.plot(x, cs(x), ':o', label='Cubic Spline', color='green')
x = np.linspace(0, 1, 100)
plt.plot(x, y(x), label='truth', color='red')
plt.legend()
```
