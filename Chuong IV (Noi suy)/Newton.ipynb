```python
import math
import numpy as np
```

```python
def newton_interpolation(x, y, target):
    n = len(x)
    y1 = [i for i in y]
    coefficients = []

    coefficients.append(y1[0])
    for i in range(n - 1):
        for j in range(n - i - 1):
            y1[j] = (y1[j + 1] - y1[j]) / (x[j + i + 1] - x[j])
        coefficients.append(y1[0])
    
    print("Các hệ số phương trình nội suy:")
    for i, coef in enumerate(coefficients):
        print(f"a[{i}] = {coef}")
    
    res = coefficients[-1]
    for i in range(n - 2, -1, -1):
        res = res * (target - x[i]) + coefficients[i]

    return res
```

```python
from decimal import Decimal, getcontext

getcontext().prec = 4

x = [Decimal(0), Decimal(0.5)]
y = [Decimal(np.exp(-0)), Decimal(np.exp(-0.5))]
res1 = newton_interpolation(x, y, Decimal(0.25))
print("=> f(0.25) =", res1)

x = [Decimal(0.5), Decimal(1)]
y = [Decimal(np.exp(-0.5)), Decimal(np.exp(-1))]
res2 = newton_interpolation(x, y, Decimal(0.25))
print("=> f(0.75) =", res2)

print()

x = [Decimal(0), Decimal(0.5), Decimal(1)]
y = [Decimal(np.exp(-0)), Decimal(np.exp(-0.5)), Decimal(np.exp(-1))]
res3 = newton_interpolation(x, y, Decimal(0.25))
res4 = newton_interpolation(x, y, Decimal(0.75))
print("Errors for f(0.25) =", abs(res3 - res1))
print("Errors for f(0.75) =", abs(res4 - res2))
```
