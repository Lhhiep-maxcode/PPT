```python
import numpy as np
from scipy.interpolate import CubicSpline
from math import prod
```

```python
def langrangeInterpolation(x: list, y: list, givenPoint):
    sz = len(x)
    res = 0
    for i in range(sz):
        lx = prod([(givenPoint - x[j]) for j in range(sz) if j != i])
        lxi = prod([(x[i] - x[j]) for j in range(sz) if j != i])
        res += lx / lxi * y[i]
    return res
```

```python
x = [8.1, 8.3, 8.6, 8.7]
y = [16.94410, 17.56492, 18.50515, 18.82091]

res = langrangeInterpolation(x, y, 8.4)
print("Nội suy Lagrange cho ý a: f(8.4) = ", res)
```

```python
x = [-0.75, -0.5, -0.25, 0]
y = [-0.0718125, -0.02475000, 0.33493750, 1.101]

res = langrangeInterpolation(x, y, -1/3)
print("Nội suy Lagrange cho ý b: f(-1/3) = ", res)
```
